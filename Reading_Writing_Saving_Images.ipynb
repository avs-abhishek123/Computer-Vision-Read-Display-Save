```python
# Python program to read image using OpenCV

# importing OpenCV(cv2) module
import cv2
from PIL import Image
from numpy import asarray
# Save image in set directory
# Read RGB image
img_in_np_array_format_using_opencv = cv2.imread('/content/elon.jpg')

#OpenCV uses BGR as its default colour order for images, matplotlib uses RGB. 
#When you display an image loaded with OpenCv in matplotlib the channels will be back to front.
#The easiest way of fixing this is to use OpenCV to explicitly convert it back to RGB, much like you do when creating the grayscale image.

RGB_img = cv2.cvtColor(img_in_np_array_format_using_opencv, cv2.COLOR_BGR2RGB)


# Below is the way of creating Pillow 
# image from our numpyarray
pilImage = Image.fromarray(RGB_img)
print(pilImage)
print(type(pilImage))
pilImage
```

# 2. Using MatplotLib :

* Matplotlib comes with a wide variety of plots.

```python
# Python program to read
# image using matplotlib

# importing matplotlib modules
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Read Images
img_in_np_array_format_using_matplotib_mpimg_imread = mpimg.imread('/content/elon.jpg')
print(type(img_in_np_array_format_using_matplotib_mpimg_imread))
# Output Images
plt.imshow(img_in_np_array_format_using_matplotib_mpimg_imread)
#plt.axis("OFF")
plt.show()

plt.imshow(img_in_np_array_format_using_matplotib_mpimg_imread)
#plt.axis("OFF")
plt.show()


plt.imshow(img_in_np_array_format_using_matplotib_mpimg_imread)
plt.axis("OFF")
plt.show()
```

## Loading the images via Pillow Library

PIL has **open()** function which opens up an image and digital file format can be retrieved as well as pixel format.

```python
from PIL import Image
image_in_PILImage_using_PIL = Image.open('/content/elon.jpg')
  
# summarize some details about the image
print(image_in_PILImage_using_PIL.format)
print(image_in_PILImage_using_PIL.size)
print(image_in_PILImage_using_PIL.mode)
image_in_PILImage_using_PIL
```
